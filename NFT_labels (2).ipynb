```python
import collections
def encoding1(dataframe):#encode column Category
  cate=collections.OrderedDict()
  indx=0
  dataframe=dataframe[dataframe['Category'].notnull()]
  for value in dataframe['Category']:
    if value not in cate:
      cate[value]=indx
      indx+=1
  cate=dict(cate)
  return cate
  
  # print(dataframe['Category'])
```

```python
cate1=encoding1(labels)
print(cate1)
```

```python
labels['Category']=labels['Category'].replace(cate1)
```

```python
def cleaning1(dataframe):
  dataframe['MM/YYYY']=dataframe['Released']
  res= dataframe['Released'].apply(lambda x: pd.Series(x.split('/')))#split MM/YYYY by '/'
  res.rename(columns={0:'MM',1:'YY'},inplace=True)
  res = res[['MM','YY']]
  dataframe['MM']=res['MM']
  dataframe['YY']=res['YY']
  dataframe['Attributes']=dataframe['Attributes'].str.split(pat=",")#split Attributes by ','
  return dataframe
```

```python
cleaning1(labels)
```
